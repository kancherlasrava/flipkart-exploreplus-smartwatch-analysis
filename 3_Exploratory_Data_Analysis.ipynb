```python
Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[(df["Price"] < lower) | (df["Price"] > upper)]
```

### Strap Color vs Brand (Heatmap)

Analyze the distribution of strap colors across different brands to identify dominant color preferences and brand-specific design patterns.

```python
ct = pd.crosstab(df['Strap_color'],df['Brand']) 
plt.figure(figsize=(12,6)) 
sns.heatmap(ct, cmap="Blues", annot=False) 
plt.title('Strap Color vs Brand (Heatmap)')
plt.show()
```

**Insight:**  
The heatmap shows that neutral colors such as Black, Silver, and Blue dominate across most brands. 
Certain brands exhibit higher concentration in specific color categories, indicating standardized design strategies rather than wide color diversification.
